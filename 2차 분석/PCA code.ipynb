```python
import pandas as pd
```

```python
data = pd.read_csv('C:/Users/jykan/Desktop/창원_데이터셋/final_data.csv', sep = ',')
data.drop(['Unnamed: 0', 't_dust', 'dust_airkorea'], axis = 1, inplace = True)
```

```python
def remove_comma(x):
    return x.replace(',', '')
data['p_total'] = data['p_total'].apply(remove_comma)
data['young_old'] = data['young_old'].apply(remove_comma)
data
```

```python
data['p_total'] = pd.to_numeric(data['p_total'])
data['young_old'] = pd.to_numeric(data['young_old'])
```

```python
new = data[['address', 'traffic/area', 'max_temp', 'dust_naver','f_pop/area', 'young_old/area']]
```

```python
new
```

```python
# 주성분 분석
from sklearn.decomposition import PCA
import matplotlib
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

matplotlib.rc('font', family = 'Malgun Gothic')
```

```python
# X 데이터, Y 데이터 분할
df_raw_x = new.drop('address', axis = 1)
df_raw_y = new[['address']]
```

```python
# 변수 이름 저장
x_cols = df_raw_x.columns

# Scaling
scaler = StandardScaler()
nd_scaled = scaler.fit_transform(df_raw_x)

# DataFrame으로 저장
df_scaled = pd.DataFrame(nd_scaled, columns = x_cols)

# 주성분 분석
pca = PCA(n_components = 1)
nd_pca = pca.fit_transform(df_scaled)

# 생성된 주성분 점수 확인
print("PCA Data : \n {}".format(nd_pca))
print("Shape : \n {}".format(nd_pca.shape))
```

```python
# 주성분 점수 저장
df_pca = pd.DataFrame(nd_pca, columns = ['Prin1'])

# 주성분 분석용 데이터와 주성분 점수 결합
df_scaled_pca = df_scaled.join(df_pca)
df_scaled_pca.head()
```

```python
pca = pd.DataFrame(nd_pca, columns = ['PCA'])
pca
```

```python
new_pca = pd.merge(new, pca, right_index = True, left_index = True)
new_pca.sort_values(by = "PCA", ascending = False, inplace = True)
```
