# DBSCAN (Clustering)

- 밀도 기반의 군집분석 (Density-Based Spatial Clustering of Applications with Noise)
- 어느 점을 기준으로 반경 x내에 점이 n개 이상 있으면 동일 군집으로 인식하는 방식
- 장점 : K-평균 방법과는 달리 군집의 수를 지정할 필요가 없음. 밀도에 따라 군집을 연결하므로 기하학적 형태의 데이터도 분류 가능. 이상치 탐색 가능.
- 단점 : 많은 연산으로 인해 속도가 느림.

```python
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn.cluster import DBSCAN
```

```python
final.columns
```

```python
new_df = final[['address', 'traffic/area', 'max_temp', 'dust_naver','f_pop/area', 'young_old/area']]
```

```python
x = new_df.iloc[:, 1:]
y = new_df.iloc[:,0]
```

```python
dbscan = DBSCAN(eps = 1.4, min_samples = 3).fit(x)
cluster_dbscan = dbscan.labels_
for(i, val) in enumerate(cluster_dbscan):
    if val < 0 :
        cluster_dbscan[i] = 0
    else:
        cluster_dbscan[i] = cluster_dbscan[i] + 1
x["CLusterDBScan"]  = cluster_dbscan
x
```

- 결과가 좋지 않음을 알 수 있다.

## 1) standard scaler

```python
# 데이터 표준화
scaler = StandardScaler()
scaled_x = scaler.fit_transform(x)
```

```python
dbscan = DBSCAN(eps = 1.0, min_samples = 4).fit(scaled_x)
cluster_dbscan = dbscan.labels_
for(i, val) in enumerate(cluster_dbscan):
    if val < 0 :
        cluster_dbscan[i] = 0
    else:
        cluster_dbscan[i] = cluster_dbscan[i] + 1
new_df["CLusterDBScan"]  = cluster_dbscan
```
