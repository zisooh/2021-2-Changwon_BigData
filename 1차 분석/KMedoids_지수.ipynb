```python
# 최적 클러스터 개수 결정
# 1) 엘보우 기법
def elbow(X) :
    wcss = []
    for i in range(1,11):
        model = KMedoids(n_clusters = i, init = "k-medoids++", random_state = 0)
        model.fit(X)
        wcss.append(model.inertia_)
    plt.figure(figsize=(10,10))
    plt.plot(range(1,11), wcss, marker='o')
    plt.xlabel('Number of clusters')
    plt.ylabel('WCSS')
    plt.show()
```

## MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df_o)
```

```python
df1 = pd.DataFrame(scaled)
df1.columns = df_o.columns
df1.head()
```

```python
elbow(df1)
```

```python
pca = PCA(2)
data = pca.fit_transform(df1)
```

```python
cobj = KMedoids(n_clusters=6).fit(data)
label = cobj.labels_
centers = np.array(cobj.cluster_centers_)
plt.figure(figsize=(7,7))
uniq = np.unique(label)
for i in uniq:
   plt.scatter(data[label == i , 0] , data[label == i , 1] , label = i)
plt.scatter(centers[:,0], centers[:,1], marker="x", color='k')
#This is done to find the centroid for each clusters.
plt.legend()
plt.show()
```

```python
cobj.inertia_
```
